```python
### importing libaries###
import pandas as pd 
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import numpy as np
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.preprocessing import StandardScaler
```

```python
### filled LotFrontage with the median because missing values represent unknown measurements, not absence###
### Filled missing LotFrontage with median ###
df['LotFrontage'] = df['LotFrontage'].fillna(df['LotFrontage'].median())
```

```python
df.select_dtypes(include=["int64", "float64"]).columns
```

```python
 ##Using  only key numeric features (no Id, no categorical for now)
numeric_features = ['LotFrontage', 'LotArea', 'OverallQual', 'OverallCond', 
                    'YearBuilt', 'YearRemodAdd', 'MasVnrArea', 'BsmtFinSF1',
                    'TotalBsmtSF', 'GrLivArea', 'FullBath', 'BedroomAbvGr', 
                    'TotRmsAbvGrd', 'GarageArea']
```

```python
X = df[numeric_features]
y = df['SalePrice']
X = X.fillna(X.median())
print(X.isnull().sum().sum())
```

```python
### STEP 6: SPLIT DATA INTO TRAIN & TEST SETS ###
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2,)
```

### TAINING AND LEARNING AND TESTING DATA AND REALITY CHECK ###

```python
### train linear regression model (baseline) ###
model = LinearRegression()
model.fit(X_train, y_train)
print("Model trained successfully!")
```

```python
### making predictions ###
y_pred = model.predict(X_test)
```

```python
### evaluating model performace ###
np.sqrt(mean_squared_error(y_test, y_pred))
r2 = r2_score(y_test, y_pred)

print(f"RMSE: ${np.sqrt(mean_squared_error(y_test, y_pred)):,.2f}")
print(f"R²: {r2_score(y_test, y_pred):.4f}")
```

```python
### errors analysis step ###

errors = y_test - y_pred

print(errors.describe())

#bias (mean)##
##spread (std)##
##extreme failures## 
```

```python
 ### interpreting model and coeficients ###
coef_df = pd.DataFrame({
    "Feature": X_train.columns,
    "Coefficient": model.coef_
})

coef_df = coef_df.sort_values(by="Coefficient", ascending=False)
coef_df
```

### feature scaling ###

```python
### training linear regression on a scaled data ###
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)


model_scaled = LinearRegression()
model_scaled.fit(X_train_scaled, y_train)

y_pred_scaled = model_scaled.predict(X_test_scaled)


rmse_scaled = np.sqrt(mean_squared_error(y_test, y_pred_scaled))
r2_scaled = r2_score(y_test, y_pred_scaled)

print(f"Scaled RMSE: ${rmse_scaled:,.2f}")
print(f"Scaled R²: {r2_scaled:.4f}")
```
